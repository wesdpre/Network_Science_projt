```python
df1 = pd.read_csv("Flights_norm.csv")
df1
```

```python
df1_sorted = df1.sort_values(by="betweenesscentrality", ascending=False)
df1_sorted.head(6)
```

```python
df1_sorted.head(30)
```

```python
df1_sorted2 = df1_sorted.sort_values(by="outdegree", ascending=False)
outdegree_desc = df1_sorted2.reset_index(drop=True)
outdegree_desc[outdegree_desc["name"] == "Ted Stevens Anchorage International Airport" ] # ranking outdegree = 279, BC = 5
outdegree_desc[outdegree_desc["name"] == "Guarulhos - Governador André Franco Montoro International Airport" ] # ranking outdegree = 74, BC = 11
outdegree_desc[outdegree_desc["name"] == "Sydney Kingsford Smith International Airport" ] # ranking outdegree = 85, BC = 14
outdegree_desc[outdegree_desc["name"] == "Brisbane International Airport" ] # ranking outdegree = 145, BC = 16
```

```python
df1_sorted2 = df1_sorted.sort_values(by="outdegree", ascending=False)
```

```python
# Step 1: Reset index so we can get rankings
df1_ranked = df1_sorted.reset_index(drop=True)
df2_ranked = df1_sorted2.reset_index(drop=True)

# Step 2: Add rankings to each DataFrame
df1_ranked['bc_rank'] = df1_ranked.index +1
df2_ranked['outdegree_rank'] = df2_ranked.index +1

# Step 3: Merge on 'name' to align the data
merged = pd.merge(
    df1_ranked[['name', 'city', 'bc_rank']],
    df2_ranked[['name', 'outdegree_rank']],
    on='name'
)

# Step 4: Calculate the absolute difference in rankings
merged['rank_diff'] = (merged['bc_rank'] - merged['outdegree_rank']).abs()

# Step 5: Sort by difference
final = merged.sort_values(by='bc_rank', ascending=True)

# (Optional) Reset index for neatness
final = final.reset_index(drop=True)

df_final = pd.DataFrame(final)
# Show the top rows
df_final.head(50)
```
